## Mini Project Idea 2: Scrape Book Titles & Prices

We’ll use http://books.toscrape.com/
 (made for scraping practice).

Your Task

Go to the homepage (http://books.toscrape.com/).

For each book on the first page, scrape:


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

all_books = []

print("Book Title  ||  Price  ||  Availability")
for i in range(1,51):
    if i==1:
        url = "https://books.toscrape.com/"
    else:
        url = f"https://books.toscrape.com/catalogue/page-{i}.html"

    response = requests.get(url)

    soup = BeautifulSoup(response.text,"html.parser")

    books = soup.find_all("article",class_="product_pod")
    for book in books:
        book_title = book.h3.a["title"]
        Price = book.find("p",class_="price_color").get_text()
        Availability = book.find("p",class_="instock availability").get_text(strip=True)
        all_books.append(
            {
                "Book_Title":book_title,
                "Price" : Price,
                "Availability":Availability
            }
        )
        print(f"{book_title}  ||  {Price}  ||  {Availability}")
        

print("Data Stored..")

df = pd.DataFrame(all_books)

df.to_csv("books.csv",index=False)
print("CSV File stored successfully..")
```
